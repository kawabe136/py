```python
import numpy as np
from numpy.random import randn
import pandas as pd

# scipyの統計パッケージも使います。
from scipy import stats

# 描画のためのライブラリです。
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns

# ブラウザ内に画像を埋め込むための命令です
%matplotlib inline

from statistics import mean, median,variance,stdev
```

```python
dframe = pd.read_excel('hist.xlsx', sheetname='Sheet1')
```

```python
data = dframe['Data'].values
```

```python
# ヒストグラム　ファイル出力
cal_mean = mean(data)
cal_median = median(data)
cal_variance = variance(data)
cal_stdev = stdev(data)
cal_max = max(data)
cal_min = min(data)
cal_n = len(data)

plt.hist(data, bins=30, ec='black') #edge color
plt.title("histgram n={}, mean={:.3f}, sigma={:.3f}, max={}, min={}" .format(cal_n, cal_mean, cal_stdev, cal_max, cal_min))
plt.xlabel('x')
plt.ylabel('y')
# plt.grid(which='major',color='black',linestyle='-')
# plt.grid(which='minor',color='black',linestyle='-')
# plt.grid(color='gray')
plt.savefig('figure.png')
```

```python
# 累積値を出力
cal_mean = mean(data)
cal_median = median(data)
cal_variance = variance(data)
cal_stdev = stdev(data)
cal_max = max(data)
cal_min = min(data)
cal_n = len(data)

plt.hist(data, bins=30, ec='black', cumulative=True) #edge color
plt.title("histgram n={}, mean={:.3f}, sigma={:.3f}, max={}, min={}" .format(cal_n, cal_mean, cal_stdev, cal_max, cal_min))
plt.xlabel('x')
plt.ylabel('y')
```

```python
# 累積値を出力
cal_mean = mean(data)
cal_median = median(data)
cal_variance = variance(data)
cal_stdev = stdev(data)
cal_max = max(data)
cal_min = min(data)
cal_n = len(data)

plt.hist(data, bins=cal_n, ec='black', normed=True, cumulative=True, histtype="step", range=(0, 4)) #edge color
plt.title("histgram n={}, mean={:.3f}, sigma={:.3f}, max={}, min={}" .format(cal_n, cal_mean, cal_stdev, cal_max, cal_min))
plt.xlabel('x')
plt.ylabel('y')
```
